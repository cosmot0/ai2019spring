## Basic data types

```python
x = 3
print(type(x)) # Prints "<class 'int'>"
print(x)       # Prints "3"
print(x + 1)   # Addition; prints "4"
print(x - 1)   # Subtraction; prints "2"
print(x * 2)   # Multiplication; prints "6"
print(x ** 2)  # Exponentiation; prints "9"
x += 1
print(x)  # Prints "4"
x *= 2
print(x)  # Prints "8"
```

```python
y = 2.5
print(type(y)) # Prints "<class 'float'>"
print(y, y + 1, y * 2, y ** 2) # Prints "2.5 3.5 5.0 6.25"
```

## Operators

### + operator

```python
print((1, 2, 3) + (4, 5, 6))
print([1, 2, 3] + [4, 5, 6])
print("Hello" + " " + "World")
```

### * operator
The * operator produces a new tuple, list, or string that "repeats" the original content.

```python
print((1, 2, 3) * 3)
print([1, 2, 3] * 3)
print("Hello "*3)
```

## Containers
Python includes several built-in container types: lists, dictionaries, sets, and tuples.

### Tuple
A simple immutable (변경할 수 없는, 불변의) ordered sequence of items

```python

# creating a tuple
months = ('January','February','March','April','May','June','July','August','September','October','November','December')

print(months[0])
print("index of 7 ==> " , months[7])

for i in range(12):
  print(months[i])
```

To print one by one

```python
# iterate through them:
for i in range(12):
  print(months[i])
```

```python
t = ('john', 32, (2,3,4,5), 'hello') #소괄호 변경x
print(t)

print(t[2])
print(t[2][1])
print(t[:2]) # index NOT included #자기는빼고 앞에꺼까지
print(t[2:]) # index included O #두번째부터 끝까지

print(t[-1])
print(t[-2])
```

### List
Mutable (바꿀수 있는, 변경가능한) ordered sequence of items of mixed types

```python
li = ['hallym', 1, 3.141572, 'hello']
print(li)
li[1] = 45
print(li)
li.append('September') #append 뒤에붙이다
print(li)
```
