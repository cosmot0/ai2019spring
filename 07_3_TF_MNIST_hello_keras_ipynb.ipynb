```
import tensorflow as tf
mnist = tf.keras.datasets.mnist
```

```
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

계산하기 편하려고 255로 나눔

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  #tf.keras.layers.Dense(512, activation=tf.nn.relu), #히든레이어 512개 추가
  #tf.keras.layers.Dense(256, activation=tf.nn.relu)  
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

모델의 구조

텐서플로우에서는 784개(28*28)의 데이터를 한줄(한층)로 쭉 나열해서 저장
Dense : 거미줄,
케라스는 layer로 구성  
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(512, activation=tf.nn.relu), #히든레이어 512개 추가
  
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
]) <가장중요한 파트>


```

model.fit(x_train, y_train, epochs=5) #사람이 정해준(hyperparameters) 5 epoch
```

#### Test accuarcy
* testset을 가지고 돌려본 accuracy (안봤던data)

```
model.evaluate(x_test, y_test)
```

#### Training accuarcy
* train dataset을 가지고 돌려본 accuracy(봤던data)

```
model.evaluate(x_train, y_train)
```
