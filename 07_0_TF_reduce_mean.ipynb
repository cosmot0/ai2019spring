### About reduce_mean 

The functionality of numpy.mean and tensorflow.reduce_mean are the *same*

```
import numpy as np
import tensorflow as tf
```

```
c = np.array([[3.,4], [5.,6], [6.,7]])
print(c)
```

```
m1 = np.mean(c, axis = 1) #mean을 구할때 axis를 지나지 않고 구하면 전체를 구함
print(m1)
```

```
m2 = tf.reduce_mean(c,axis = 1) #reduce_mean 평균을 구하는것
with tf.Session() as sess:
    result = sess.run(m2)
    print(result)
```
